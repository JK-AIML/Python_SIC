```python
year = 2100
if ( ((year % 4 ==0) and (year % 100 !=0)) or (year %400 ==0) ):
    print(f"yes {year} is a leap year")
else:
    print("Not a leap year")
```

```python
list_prime = []
list_divisible = []
for i in range(2, 20):
    for j in range(2, i):
        if i%j ==  0:
            list_divisible.append(i)
            break
    else:
        list_prime.append(i)

print(f"Divisible is {list_divisible}")
print(f"Prime is: {list_prime}")
```

```python
number_list = input().split(' ')
jackpot_list = ["3", "8", "9"]
n = 0
for i in number_list:
    for j in jackpot_list:
        if i == j:
            jackpot_list.remove(j)
            n+=1;
print(f"n: {n} matched")

#OR
n = 0
for i in number_list:
    if i in jackpot_list:
        jackpot_list.remove(i)
        n+=1
print(f"n: {n} matched")
```
