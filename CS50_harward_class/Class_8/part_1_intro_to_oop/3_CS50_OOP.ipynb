```python
#when we use list/tuples, we often can make mistakes when using indexes so it may be better to use dicts
def main4():
    student = get_student()
    for i in student:
        if i == "Harry":
            student[i] = "Gryffindor"
        print(f"{i} is in {student[i]} house")

def get_student():
    name = input("Enter Name: ")
    house = input("Enter House: ")
    return {name: house}
main4()
```

```python
#work around using dict:
def main5():
    students = []
    student = get_student()
    students = students + [student]
    for i in students:
        if i[0] == "Harry":
            i[1] = "Gryffindor"
        print(f"{i[0]} is in {i[1]} house")

def get_student():
    name = input("Enter Name: ")
    house = input("Enter House: ")
    return [name,house]
main5()
```

```python
#we still use interations to get key names, how to go around that? use keys as column names instead of the data itself
#we want to use custom indexes i.e. keys (like we use index for lists)
# this is standard as key names shoud be unique and easy to asscess

def main6():
    student = get_student()
    if student["name"] == "Harry":
        student["house"] = "Gryffindor"
    print(f"{student['name']} is in {student['house']} house")

def get_student():
    name = input("Enter Name: ")
    house = input("Enter House: ")
    return {"name": name, "house": house}

main6()
```
